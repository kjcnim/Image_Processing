```
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp',cv2.IMREAD_GRAYSCALE)

#image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image, cmap='gray')
```

```
#여기서는 측정되는 value 값들에 무엇들이 있는지 구한다. 또한 value들의 개수를 이용하여 list size를 초기화 할 것이다.
value = []

for i in range(512):
  for j in range(512):
    v = image[i][j]
    if v not in value:
      value.extend([v]) #여기서 x좌표의 값을 다 찾는다.

n = len(value)
value.sort() #그래프를 그리기 때문에 순서대로 정렬한다
print(value)
```

```
#여기서는 p list에 해당되는 값이 몇 개있는지를 확인하는 것이다.
#따라서 count_list의 순서와 p list의 순서는 연관이 있다.
count = 0
count_list = []

for k in range(n):
  z = value[k]
  for i in range(512):
    for j in range(512):
      u = image[i][j]
      if z == u:
        count = count+1
  count_list.extend([count])
  count = 0 #초기화

print(count_list)
```

```
graph = np.arange(n)

plt.bar(graph,count_list)
plt.xticks(graph,value)
plt.show()
```
