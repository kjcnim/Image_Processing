```
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp',cv2.IMREAD_GRAYSCALE)

#image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image, cmap='gray')
```

```
#  value    .  value   list size   .
value = []

for i in range(512):
  for j in range(512):
    v = image[i][j]
    if v not in value:
      value.extend([v]) # x   .

n = len(value)
value.sort() #    
print(value)
```

```
# p list      .
# count_list  p list   .
count = 0
count_list = []

for k in range(n):
  z = value[k]
  for i in range(512):
    for j in range(512):
      u = image[i][j]
      if z == u:
        count = count+1
  count_list.extend([count])
  count = 0 #

print(count_list)
```

```
graph = np.arange(n)

plt.bar(graph,count_list)
plt.xticks(graph,value)
plt.show()
```
