```
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp')

image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.title('original')
plt.imshow(image)

print(image)
```

```
a = image #copy

offset = 50 #255   0 .

for i in range(512) :
  for j in range(512) :
    a[i][j][0] = image[i][j][0] + offset
    a[i][j][1] = image[i][j][1] + offset
    a[i][j][2] = image[i][j][2] + offset

print(a.shape)
print('-----------------------------------')
print(a)
print('-----------------------------------')
#print(image)    
plt.imshow(a)
```
