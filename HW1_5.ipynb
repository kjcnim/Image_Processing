```
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2

image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp',cv2.IMREAD_GRAYSCALE)

plt.title('original')
plt.imshow(image,cmap ="gray")
```

```
ret, th = cv2.threshold(image, 0, 512, cv2.THRESH_BINARY+cv2.THRESH_OTSU) #otsu 

images = [image, 0, th]

plt.titles = ["Otsu's Thresholding"]

plt.imshow(images[2],'gray')
```
