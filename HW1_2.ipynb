```
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/lena.bmp')

image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.title('original')
plt.imshow(image)

#print(image)
```

```
a = image #copy

offset = 50 #255넘어가면 그 차액만큼 0부터 더해진다.

for i in range(512) :
  for j in range(512) :
    a[i][j][0] = image[i][j][0] + offset
    a[i][j][1] = image[i][j][1] + offset
    a[i][j][2] = image[i][j][2] + offset

#print(a.shape)
#print('-----------------------------------')
#print(a) #이걸 하면 위에서 print(image)값에서 offset을 더한 값을 확인할 수 있게 된다.
#print('-----------------------------------')
#print(image)    
plt.imshow(a)
```
